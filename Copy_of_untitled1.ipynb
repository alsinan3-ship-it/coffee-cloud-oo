```python
class Task:
    def __init__(self, description):
        self.description = description


class Recommendation:
    def __init__(self, agent, message):
        self.agent = agent
        self.message = message


class Agent:
    def __init__(self, name):
        self.name = name

    def give_advice(self, task):
        return Recommendation(self.name, "No advice")

class FinanceAgent(Agent):
    def give_advice(self, task):
        return Recommendation(self.name, "Be careful with spending.")


class MarketingAgent(Agent):
    def give_advice(self, task):
        return Recommendation(self.name, "Offer discounts to increase sales.")


class OrderAgent(Agent):
    def give_advice(self, task):
        return Recommendation(self.name, "Prepare for more customer orders.")


class Founder:
    def __init__(self, name):
        self.name = name

    def make_decision(self, recommendations):
        print("\nRecommendations:")
        for r in recommendations:
            print(r.agent + ":", r.message)

        print("\nFinal Decision:")
        print("Apply a small discount and improve order system.")


task = Task("Increase sales")

founder = Founder("AlrWas")

agents = [
    FinanceAgent("Finance Agent"),
    MarketingAgent("Marketing Agent"),
    OrderAgent("Order Agent")
]

recommendations = []

for agent in agents:
    rec = agent.give_advice(task)
    recommendations.append(rec)

founder.make_decision(recommendations)
```
